```python
konh = pd.read_csv('/content/konh.csv')
```

```python
labels_form = konh.TARGET
```

```python
def n05(x):
    if x>=0.5:
        return 1
    return 0
```

```python
konh['TARGET'] = labels_form.apply(n05)
```

```python
konh = konh.drop(['Unnamed: 0'],axis=1)
```

```python
N_THREADS = mp.cpu_count() # threads cnt for lgbm and linear models
N_FOLDS = 10 # folds cnt for AutoML - 10 is better
RANDOM_STATE = 42 # fixed random state for various reasons
TIMEOUT = 3600 # Time in seconds for automl run
TARGET_NAME = 'TARGET' # Target column name
```

```python
%%time

roles = {'target': TARGET_NAME,
         'drop': ['id_contract','id_client']
         }
```

```python
%%time

task = Task('binary', loss='logloss', metric='auc')
```

```python
#### %%time 
automl = TabularAutoML(task = task, 
                       timeout = 1800,
                       cpu_limit = N_THREADS,
                       #general_params = {'use_algos': [['linear_l2', 'lgb', 'lgb_tuned']]},
                       reader_params = {'n_jobs': N_THREADS, 'cv': N_FOLDS, 'random_state': RANDOM_STATE},
                      )
oof_pred = automl.fit_predict(konh, roles = roles)
logging.info('oof_pred:\n{}\nShape = {}'.format(oof_pred, oof_pred.shape))
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(konh.TARGET,oof_pred.data)
```
