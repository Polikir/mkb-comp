```python
time_features = ['SIGN_DATE', 'DATEFIRSTREG', 'TAXREG_REGDATE', 'TAXREGPAY_REGDATE', 'BIRTHDATE']
```

```python
# 01JAN2018:00:00:00 - date format
from datetime import datetime
import math

def parse_date(date):
    try:
        monthes = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
        day = int(date[0:2])
        month = monthes.index(date[2:5]) + 1
        year = int(date[5:9])
        return datetime(year, month, day)
    except:
        pass
```

```python
order = {'0 .. 5':0, '6 .. 10':1, '11 .. 15':2, '16 .. 50':3, '51 .. 100':4,
        '101 .. 150':5, '151 .. 200':6, '201 .. 250':7, '251 .. 500': 8,
        '501 .. 1000':6, '1001 .. 5000':7, '5000 и более':8}

def ord_encoding(wrange):
    try:
        return order[wrange]
    except:
        pass
```

```python
train['WORKERSRANGE'] = train.apply(lambda row: ord_encoding(row['WORKERSRANGE']), axis = 1)
```

```python
test['WORKERSRANGE'] = test.apply(lambda row: ord_encoding(row['WORKERSRANGE']), axis = 1)
```

```python
"""OKATO_FED AND OKTMO_FED - DOUBLE EACH OTHER"""

cats_to_drop = ['OKTMO_FED', 'OKTMO_CODE']
train = train.drop(columns = cats_to_drop)
test = test.drop(columns = cats_to_drop)
```

```python
cat_features.remove('OKTMO_FED')
cat_features.remove('OKTMO_CODE')
```

```python
from sklearn.model_selection import train_test_split
from feature_engine.encoding import WoEEncoder, RareLabelEncoder


data = train.copy()

# Separate into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    data.drop(columns=[''])
    data['TARGET'], test_size=0.3, random_state=0)

# set up a rare label encoder
rare_encoder = RareLabelEncoder(tol=0.03, n_categories=10, variables=cat_features)

# fit and transform data
train_t = rare_encoder.fit_transform(X_train)
test_t = rare_encoder.transform(X_train)

# set up a weight of evidence encoder
woe_encoder = WoEEncoder(variables=cat_features)

# fit the encoder
woe_encoder.fit(train_t, y_train)

# transform
train_t = woe_encoder.transform(train_t)
test_t = woe_encoder.transform(test_t)
```
